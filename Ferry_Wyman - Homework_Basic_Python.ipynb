** DATA CLEANSING - Remove unwanted characters ('+' and 'FREE') in Installs Columns

```python
gdrop['Installs'] = gdrop['Installs'].map(lambda x: x.lstrip('+-').rstrip('FfRrEe'))
```

```python
gdrop['Installs'] = gdrop['Installs'].str.replace(r'\D', '')
```

```python
gdrop.head (5)
```

** DATA MANIPULATION - Change 'Intalls' column type from object to numeric

```python
gdrop["Installs"] = pd.to_numeric(gdrop["Installs"])
```

```python
gdrop.info()
```

** Sorting table by Installs App and Group it By Category column

```python
gdrop = gdrop.sort_values(by ='Installs' , ascending = False)
```

```python
cat = gdrop.groupby('Category')
```

```python
gdrop.head (3)
```

** Whats App Messenger, Google Play Books, and Subway Surfers are top 3 App based on Installs

```python
for Category,Category_gdrop in cat :
    print(Category)
    print()
    print(Category_gdrop[['App','Installs','Rating']].sort_values(by ='Installs' , ascending = False).head(20))
    print()
```

** DATA MANIPULATION - Change 'Reviews' column type from object to numeric, replace 'K' 'M' with int)

```python
gdrop.Reviews = (gdrop.Reviews.replace(r'[KM]+$', '', regex=True).astype(float) * \
                 gdrop.Reviews.str.extract(r'[\d\.]+([KM]+)', expand=False)
                 .fillna(1)
                 .replace(['K','M'], [10**3, 10**6]).astype(int))
```
